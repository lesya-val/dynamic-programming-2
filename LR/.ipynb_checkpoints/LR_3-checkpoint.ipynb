# Лабораторная работа №3

### Импорт библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, accuracy_score, mean_squared_error
from sklearn.inspection import permutation_importance
```

### Загрузка данных временного ряда (Максимальная цена за торговый день по акциям TESLA на ежедневном уровне)

```python
data = pd.read_csv("TSLA.csv", sep=';', parse_dates=True, index_col="Date", dayfirst=True)
aapl = data["Value"]
```

### Подготовка данных

```python
def create_features(data, look_back=60):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back)])
        y.append(data[i + look_back])
    
    # Преобразуем в массив NumPy
    X = np.array(X)
    y = np.array(y)

    # Убираем лишнюю размерность из X
    X = X.reshape(X.shape[0], X.shape[1])  # Преобразуем X в двумерный массив

    # Добавляем дополнительные признаки: скользящие средние и разности
    moving_avg_5 = np.convolve(data.flatten(), np.ones(5) / 5, mode='valid')
    moving_avg_10 = np.convolve(data.flatten(), np.ones(10) / 10, mode='valid')
    
    # Разности цен
    price_diff = np.diff(data.flatten(), n=1)
    
    # Стандартное отклонение
    rolling_std_5 = pd.Series(data.flatten()).rolling(window=5).std().to_numpy()
    rolling_std_10 = pd.Series(data.flatten()).rolling(window=10).std().to_numpy()

    # Убедимся, что размеры совпадают
    min_length = min(len(X), len(moving_avg_5), len(moving_avg_10), len(price_diff), len(rolling_std_5), len(rolling_std_10))
    
    X = X[:min_length]
    moving_avg_5 = moving_avg_5[:min_length]
    moving_avg_10 = moving_avg_10[:min_length]
    price_diff = price_diff[:min_length]
    rolling_std_5 = rolling_std_5[:min_length]
    rolling_std_10 = rolling_std_10[:min_length]

    # Объединяем все признаки
    X_extended = np.column_stack((X, 
                                   moving_avg_5[:len(X)], 
                                   moving_avg_10[:len(X)], 
                                   price_diff[:len(X)], 
                                   rolling_std_5[:len(X)], 
                                   rolling_std_10[:len(X)]))
    
    return X_extended, y[:min_length]  # Убираем лишние элементы из y для соответствия по размерности
```

### Нормализация и создание признаков

```python
# Масштабируем данные до разделения на train/test
scaler = MinMaxScaler()
aapl_scaled = scaler.fit_transform(aapl.values.reshape(-1, 1))

# Создаём признаки и целевые значения
X, y = create_features(aapl_scaled)

# Нормализуем целевую переменную
y_scaled = scaler.fit_transform(y.reshape(-1, 1)).ravel()
```

### Разделение данных на обучающую и тестовую выборки

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_scaled, test_size=0.2, shuffle=False)

# Create a mask for non-NaN values in X_train
mask_train = ~np.isnan(X_train).any(axis=1)

# Apply the mask to both X_train and y_train
X_train = X_train[mask_train]
y_train = y_train[mask_train]

# Create a mask for non-NaN values in X_test
mask_test = ~np.isnan(X_test).any(axis=1)

# Apply the mask to both X_test and y_test
X_test = X_test[mask_test]
y_test = y_test[mask_test]
```

### Гиперпараметры для моделей

```python
# Logistic Regression
lr_param_grid = {
    'penalty': ['l2'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'saga'],
    'C': [0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 200]
}

# SVR
svr_param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'kernel': ['linear', 'poly'],
    'epsilon': [0.01, 0.1, 0.5, 1],
    'gamma': ['scale','auto', 0.1, 1, 10]
}

# Random Forest
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': [None, 'sqrt', 'log2'],
}

# XGBoost
xgb_param_grid = {
    'n_estimators': [100, 200, 500],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 10, 12],
    'subsample': [0.5, 0.7, 1.0],
    'colsample_bytree': [0.5, 0.7, 1.0],
    'gamma': [0, 0.1, 0.5],
}
```

### Оценка моделей

```python
def evaluate_model(model, X_test, y_test, model_name):
    X_test_reshaped = X_test.reshape(-1, X_test.shape[1])
    if model_name == 'Logistic Regression':
        # Для логистической регрессии прогнозируем направление движения
        y_pred_direction = model.predict(X_test_reshaped)
    
        # Прогнозируем цены на основе направления движения
        y_pred = np.where(y_pred_direction == 1, y_test + 0.01 * scaler.scale_[0], y_test - 0.01 * scaler.scale_[0])
    else:
        y_pred = model.predict(X_test_reshaped)
    
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    print(f'{model_name} - MAE: {mae:.4f}, RMSE: {rmse:.4f}')
    return y_pred
```

### Функция для визуализации

```python
def visualize_forecast(predictions, model_name):
    plt.figure(figsize=(12, 6))
    
    # Визуализация фактических значений
    plt.plot(aapl.index[-len(y_test):], scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual', color='blue')
    
    # Визуализация прогнозируемых значений
    plt.plot(aapl.index[-len(y_test):], scaler.inverse_transform(predictions.reshape(-1, 1)), label=model_name, color='orange')
    
    # Настройка графика
    plt.title('Сравнение прогнозов моделей машинного обучения')
    plt.xlabel('Дата')
    plt.ylabel('Цена акции')
    plt.legend()
    
    # Показать график
    plt.show()
```

### Логистическая регрессия

```python
X_train_lr = X_train[1:].reshape(-1, X_train.shape[1])
y_direction = (y_train[1:] > y_train[:-1]).astype(int).ravel()

X_train_lr = X_train_lr[~np.isnan(X_train_lr).any(axis=1)]

lr_model = LogisticRegression()
lr_model.fit(X_train_lr, y_direction)

grid_search = GridSearchCV(
    LogisticRegression(),
    lr_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_search.fit(X_train_lr, y_direction)
best_lr_model = grid_search.best_estimator_
print(f"Лучшие гиперпараметры: {grid_search.best_params_}")

best_lr_model.fit(X_train_lr, y_direction)

lr_pred = evaluate_model(best_lr_model, X_test, y_test, 'Logistic Regression')

# Визуализация результатов
visualize_forecast(lr_pred, model_name='Logistic Regression')
```

### Модель машины опорных векторов (SVR):

```python
# Преобразуем X_train и X_test в двумерные массивы
X_train_svr = X_train.reshape(X_train.shape[0], -1)  # Размер [n_samples, look_back]
X_test_svr = X_test.reshape(X_test.shape[0], -1)

# Используем GridSearchCV для настройки гиперпараметров SVR
svr_grid_search = GridSearchCV(
    SVR(),
    svr_param_grid,
    cv=5,  # число фолдов для кросс-валидации
    scoring='neg_mean_absolute_error',  # метрика для подбора параметров
    n_jobs=-1  # использование всех доступных ядер процессора
)

# Преобразуем y_train в одномерный массив
y_train_svr = y_train.ravel()

# Выполняем поиск оптимальных гиперпараметров
svr_grid_search.fit(X_train_svr, y_train_svr)

# Получаем лучшую модель и её параметры
best_svr_model = svr_grid_search.best_estimator_
best_svr_params = svr_grid_search.best_params_
print(f"Лучшие гиперпараметры для SVR: {best_svr_params}")

# Обучение модели с лучшими параметрами
best_svr_model.fit(X_train_svr, y_train_svr)

# Прогнозирование на тестовой выборке
svr_pred = evaluate_model(best_svr_model, X_test_svr, y_test, 'SVR')

# Визуализация результатов
visualize_forecast(svr_pred, model_name='SVR')
```

### Модель случайного леса  (Random Forest)

```python
# Преобразование данных в двухмерный вид
X_train_rf = X_train.reshape(X_train.shape[0], -1)
X_test_rf = X_test.reshape(X_test.shape[0], -1)

# Преобразуем y_train и y_test в одномерные массивы
y_train = y_train.ravel()
y_test = y_test.ravel()

grid_search_rf = GridSearchCV(
    RandomForestRegressor(),
    rf_param_grid,
    cv=5,  # Число фолдов для кросс-валидации
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

# Выполняем поиск оптимальных гиперпараметров
grid_search_rf.fit(X_train_rf, y_train)

# Получаем лучшую модель и её параметры
best_rf_model = grid_search_rf.best_estimator_
best_rf_params = grid_search_rf.best_params_
print(f"Лучшие гиперпараметры для Random Forest: {best_rf_params}")

# Обучение модели с лучшими параметрами
best_rf_model.fit(X_train_rf, y_train)

# Прогнозирование на тестовой выборке
rf_pred = evaluate_model(best_rf_model, X_test_rf, y_test, 'Random Forest')

# Визуализация результатов
visualize_forecast(rf_pred, model_name='Random Forest')
```

### Модель градиентного бустинга (XGBoost)

```python
# Преобразуем данные в двумерный вид
X_train_xgb = X_train.reshape(X_train.shape[0], -1)
X_test_xgb = X_test.reshape(X_test.shape[0], -1)

# Преобразование меток в 1D
y_train = y_train.ravel()
y_test = y_test.ravel()

grid_search_xgb = GridSearchCV(
    XGBRegressor(),
    xgb_param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

# Выполняем поиск оптимальных гиперпараметров
grid_search_xgb.fit(X_train_xgb, y_train)

# Получаем лучшую модель и её параметры
best_xgb_model = grid_search_xgb.best_estimator_
best_xgb_params = grid_search_xgb.best_params_
print(f"Лучшие гиперпараметры для XGBoost Predictions: {best_xgb_params}")

# Обучение модели с лучшими параметрами
best_xgb_model.fit(X_train_xgb, y_train)

# Прогнозирование на тестовой выборке
pred_xgb = evaluate_model(best_xgb_model, X_test_xgb, y_test, 'XGBoost')

# Визуализация результатов
visualize_forecast(pred_xgb, model_name='XGBoost')
```

### Сравнение результатов прогнозирования моделей

#### 1. Logistic Regression
Модель демонстрирует отличные результаты с очень низкими значениями MAE и RMSE. График, который практически совпадает с исходными данными, указывает на высокую точность модели в предсказаниях. Это делает её предпочтительным выбором для данной задачи.
#### 2. SVR (Support Vector Regression)
Модель показывает более высокие значения MAE и RMSE по сравнению с Logistic Regression, что указывает на менее точные предсказания. Однако график демонстрирует разнообразие в предсказаниях, что может быть полезно в определенных контекстах. Эта модель может быть хороша для задач с нелинейными зависимостями.
#### 3. Random Forest
Результаты модели показывают значительно более высокие значения MAE и RMSE, что указывает на плохую производительность модели в данной задаче. Горизонтальная линия на графике говорит о том, что модель не смогла уловить паттерны в данных и просто предсказывает одно и то же значение около нуля.
#### 4. XGBoost
Модель имеет аналогичные результаты с Random Forest, что также свидетельствует о низкой эффективности модели в данной задаче. Как и в случае с Random Forest, график указывает на неспособность модели делать значимые предсказания.

### Оценка важности признаков для каждой модели

```python
# Получаем коэффициенты модели
coefficients = best_lr_model.coef_[0]

# Создаем DataFrame для важности признаков
importances_lr = pd.DataFrame({
    'Feature': ['moving_avg_5', 'moving_avg_10', 'price_diff', 'rolling_std_5', 'rolling_std_10'],
    'Importance': coefficients[:5]
})

# Сортируем по важности
importances_lr = importances_lr.sort_values(by='Importance', ascending=False)

# Оценка важности признаков для SVR
result = permutation_importance(best_svr_model, X_test_svr, y_test, n_repeats=30, random_state=42)
importances_svr = pd.DataFrame({
    'Feature': ['moving_avg_5', 'moving_avg_10', 'price_diff', 'rolling_std_5', 'rolling_std_10'],
    'Importance': result.importances_mean[:5]
})
importances_svr = importances_svr.sort_values(by='Importance', ascending=False)

# Для Random Forest
importances_rf = best_rf_model.feature_importances_
importances_rf_df = pd.DataFrame({
    'Feature': ['moving_avg_5', 'moving_avg_10', 'price_diff', 'rolling_std_5', 'rolling_std_10'],
    'Importance': importances_rf[:5]
})
importances_rf_df = importances_rf_df.sort_values(by='Importance', ascending=False)

# Для XGBoost
importances_xgb = best_xgb_model.feature_importances_
importances_xgb_df = pd.DataFrame({
    'Feature': ['moving_avg_5', 'moving_avg_10', 'price_diff', 'rolling_std_5', 'rolling_std_10'],
    'Importance': importances_xgb[:5]
})
importances_xgb_df = importances_xgb_df.sort_values(by='Importance', ascending=False)

# Объединение важностей признаков для всех моделей
combined_importance = pd.concat([
    importances_lr.set_index('Feature').rename(columns={'Importance': 'Logistic Regression'}),
    importances_svr.set_index('Feature').rename(columns={'Importance': 'SVR'}),
    importances_rf_df.set_index('Feature').rename(columns={'Importance': 'Random Forest'}),
    importances_xgb_df.set_index('Feature').rename(columns={'Importance': 'XGBoost'})
], axis=1)

# Умножаем все значения важности на 1000
combined_importance *= 1000

print("Важность признаков для всех моделей (умноженная на 1000):")
print(combined_importance)

# Функция для построения гистограммы важности признаков
def plot_feature_importance(importances_df, model_name):
    plt.figure(figsize=(10, 6))
    plt.barh(importances_df['Feature'], importances_df[model_name], color='skyblue')
    plt.xlabel('Важность (умноженная на 1000)')
    plt.title(f'Важность признаков для {model_name}')
    
    # Устанавливаем пределы по оси X от 0 до 2
    plt.xlim(0, 2)  
    
    # Устанавливаем метки по оси X с шагом 0.1
    plt.xticks(np.arange(0, 2.1, step=0.1))
    
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

# Построение гистограмм для каждой модели
for model in combined_importance.columns:
    plot_feature_importance(combined_importance.reset_index(), model)
```
