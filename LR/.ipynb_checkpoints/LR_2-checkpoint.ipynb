### Загрузка данных временного ряда (Максимальная цена за торговый день по акциям TESLA на ежедневном уровне)

```python
data = pd.read_csv("TSLA.csv", sep=';', parse_dates=True, index_col="Date", dayfirst=True)
data.index = pd.DatetimeIndex(data.index)

data = data.asfreq('D')
data['Value'] = data['Value'].fillna(method='ffill')
```

### Декомпозиция временного ряда

```python
decomposition = seasonal_decompose(data, model='additive', period=365)
```

### Визуализация компонентов временного ряда

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 16))
decomposition.observed.plot(ax=ax1)
ax1.set_title('Наблюдаемый')
decomposition.trend.plot(ax=ax2)
ax2.set_title('Тренд')
decomposition.seasonal.plot(ax=ax3)
ax3.set_title('Сезонность')
decomposition.resid.plot(ax=ax4)
ax4.set_title('Остаток')
plt.tight_layout()
plt.show()
```

### Получение десезонализированного ряда

```python
deseasonalized = decomposition.observed - decomposition.seasonal
```

### Создание нового графика для сравнения

```python
plt.figure(figsize=(12, 6))
plt.plot(decomposition.observed, label='Исходный ряд', color='blue')
plt.plot(deseasonalized, label='Десезонализированный ряд', color='orange', linestyle='--')
plt.title('Сравнение исходного и десезонализированного рядов')
plt.legend()
plt.show()
```

### Построение прогноза на основе выявленных тренда и сезонности на 12 периодов вперед

```python
# Период сезонности
PERIOD_SIZE = 30
season_period = 12 * PERIOD_SIZE
data_value = data['Value']

# Декомпозиция временного ряда
decomposition = seasonal_decompose(data_value, model='additive', period=season_period)
trend_component = decomposition.trend
seasonal_component = decomposition.seasonal
residual_component = decomposition.resid

# Прогнозирование тренда
# Убираем NaN из трендовой компоненты
trend_non_nan = trend_component.dropna()

# Линейное экстраполирование тренда на 12 шагов вперед
trend_forecast = np.poly1d(np.polyfit(range(len(trend_non_nan)), trend_non_nan, 1))(range(len(trend_non_nan), len(trend_non_nan) + season_period))

# Прогноз сезонности
seasonal_forecast = seasonal_component[-season_period:].values  # последние 12 значений сезонности

# Итоговый прогноз: складываем тренд и сезонность
forecast = trend_forecast + seasonal_forecast
forecast[0] = data_value.tail(1)

# Получение последнего индекса временного ряда
last_index = data_value.index[-1]

# Индексы для прогноза (например, если у вас есть даты)
forecast_index = pd.date_range(last_index, periods=season_period + 1, freq='D')[1:]

# Визуализация прогноза
plt.figure(figsize=(16, 8))
plt.plot(data_value.iloc[-season_period:], label='Исходный ряд', color='blue')
plt.plot(forecast_index, forecast, label='Прогноз на 12 периодов вперед', color='orange', linestyle='--')
plt.title('Прогноз на основе тренда и сезонности')
plt.legend()
plt.grid(True)
plt.show()
```

### Оценка качества прогноза с помощью метрик MAE и RMSE

```python
actual_values = data_value.iloc[-season_period:]

# Метрики MAE и RMSE
mae = mean_absolute_error(actual_values, forecast)
rmse = np.sqrt(mean_squared_error(actual_values, forecast))
print("MAE:", mae)
print("RMSE:", rmse)
print("Соотношение RMSE/MAE:", rmse/mae)
```

### Тест на стационарность ряда (тест Дики-Фуллера)

```python
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')

adf_test(data)
```

Значение p-value значительно больше 0.05, мы не можем отвергнуть нулевую гипотезу H₀ о наличии единичного корня. Это означает, что временной ряд не является стационарным по результатам ADF-теста.

### Приведение ряда к стационарному виду

```python
# Установим окно равное 12
rolling_mean = data_value.rolling(window=12).mean()

# Детерминируем тренд
detrended = data_value - rolling_mean
stationary_series = detrended.dropna()

plt.figure(figsize=(16, 8))
plt.plot(data_value, label='Исходный ряд')
plt.plot(rolling_mean, label='Скользящее среднее', color='red')
plt.plot(stationary_series, label='Detrended', color='green')
plt.title('Детерминированный тренд') 
plt.legend()
plt.show()
```

### Проверка стационарности после детрендирования

```python
adf_test(stationary_series)
```

### Функция для визуализации прогноза модели

```python
def plot_forecast(series, forecast, forecast_index, title):
    forecast[0] = series.tail(1)

    # Построение графика полного ряда данных
    plt.figure(figsize=(16, 8))
    plt.plot(series, label='Исходный ряд')

    # Задаем диапазоны для отображения (например, только с 20-го по 50-й день)
    start_date = '2021-01-01'
    end_date = '2022-12-31'

    # Установка диапазона по оси X
    plt.xlim(pd.to_datetime(start_date), pd.to_datetime(end_date))
    plt.plot(forecast_index, forecast, label=f'Прогноз {title}', color='red')
    plt.title(title)
    plt.legend()
    plt.show()
```

### Функция для подбора оптимальных параметров по AIC и BIC

```python
def find_optimal_params(model_type, series, param_ranges):
    aic_values = []
    bic_values = []
    
    for params in param_ranges:
        try:
            if model_type == 'AR':
                model = AutoReg(series, lags=params).fit()
            elif model_type == 'MA':
                model = ARIMA(series, order=(0, 0, params)).fit()
            elif model_type == 'ARMA':
                model = ARIMA(series, order=(params[0], 0, params[1])).fit()
            elif model_type == 'ARIMA':
                model = ARIMA(series, order=params).fit()

            aic_values.append((params, model.aic))
            bic_values.append((params, model.bic))
        except:
            continue
    
    optimal_aic = min(aic_values, key=lambda x: x[1])
    optimal_bic = min(bic_values, key=lambda x: x[1])
    
    return optimal_aic, optimal_bic
```

### Функция для диагностики модели

```python
def plot_residuals(model, title):
    residuals = model.resid

    fig, axes = plt.subplots(1, 2, figsize=(16, 8))
    
    # График остатков модели
    axes[0].plot(residuals)
    axes[0].set_title(f'Остатки модели: {title}')

    # График ACF остатков
    plot_acf(residuals, lags=40, ax=axes[1])
    axes[1].set_title(f'ACF остатков: {title}')
    
    # Плотная компоновка
    plt.tight_layout()
    plt.show()
```
